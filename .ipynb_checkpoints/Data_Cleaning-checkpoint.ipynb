```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pickle
```

```python
def clean_data(df):
    years = [str(year) for year in range(1960,2000)]
    years.extend(['Unnamed: 63','Indicator Code','Indicator Name'])
    df.head()
    df = df.drop(columns=years)
    df = df.drop(index = df[df.sum(axis=1) == 0].index)
    df['Average %'] = df.mean(axis=1)
    cols = df.columns.tolist()
    cols = cols[0:2] + cols[-1:] + cols[2:-1]
    df = df[cols]
    return df

def create_data(file_loc):
    df = pd.read_csv(file_loc)
    df = clean_data(df)
    return df
```

```python
PrimaryCompletion = pd.read_csv('Data/Primary_Completion_Rate.csv')
```

```python
PrimaryCompletion = clean_data(PrimaryCompletion)
```

```python
PrimaryCompletion.shape
```

```python
ChildEmploymentF = pd.read_csv('Data/Child_Employment_Female.csv')
ChildEmploymentF = clean_data(ChildEmploymentF)
ChildEmploymentF.shape
```

```python
ChildEmploymentM = pd.read_csv('Data/Child_Employment_Male.csv')
ChildEmploymentM = clean_data(ChildEmploymentM)
ChildEmploymentM.shape
```

```python
GovExpend = create_data('Data/Government_Expenditure.csv')
GovExpend.shape
```

```python
PopDense = create_data('Data/Population_Density.csv')
PopDense.shape
```

```python
UrbanPop = create_data('Data/Urban_Population.csv')
UrbanPop.shape
```

```python
common1 = PrimaryCompletion[PrimaryCompletion.index.isin(ChildEmploymentM.index)]
common2 = ChildEmploymentM[ChildEmploymentM.index.isin(PrimaryCompletion.index)]
common3 = pd.merge(left=common1,right=common2,how='inner',left_on='Country Name',right_on='Country Name')
common3.head()
```

```python
sns.lmplot(x='Average %_x',y='Average %_y',data=common3,fit_reg=True)
```

```python
def plot_2(df1,df2):
    common = pd.merge(left=df1,right=df2,how='inner',left_on='Country Name',right_on='Country Name')
    print(len(common),'Observations')
    return sns.jointplot(x='Average %_x',y='Average %_y',data=common,kind='reg')
```

```python
plot_2(PrimaryCompletion,ChildEmploymentF).set_axis_labels(xlabel='Primary School Enrollment %',ylabel='Female Child Employment %')
```

```python
plot_2(PrimaryCompletion,ChildEmploymentM).set_axis_labels(xlabel='Primary School Enrollment %',ylabel='Male Child Employment %')
```

```python
plot_2(PrimaryCompletion,GovExpend).set_axis_labels(xlabel='Primary School Enrollment %',ylabel='Percentage of GDP Spent on Education')
```

```python
plot_2(PrimaryCompletion,PopDense[PopDense['Average %'] < 1000]).set_axis_labels(xlabel='Primary School Enrollment %',ylabel='Population per Sq. Km')
```

```python
plot_2(PrimaryCompletion,UrbanPop).set_axis_labels(xlabel='Primary School Enrollment %',ylabel='Urban Population %')
```

```python
def CorrMtx(df, dropDuplicates = True):

    # Your dataset is already a correlation matrix.
    # If you have a dateset where you need to include the calculation
    # of a correlation matrix, just uncomment the line below:
    # df = df.corr()

    # Exclude duplicate correlations by masking uper right values
    if dropDuplicates:    
        mask = np.zeros_like(df, dtype=np.bool)
        mask[np.triu_indices_from(mask)] = True

    # Set background color / chart style
    sns.set_style(style = 'white')

    # Set up  matplotlib figure
    f, ax = plt.subplots(figsize=(11, 9))

    # Add diverging colormap from red to blue
    cmap = sns.diverging_palette(250, 10, as_cmap=True)

    # Draw correlation plot with or without duplicates
    if dropDuplicates:
        sns.heatmap(df, mask=mask, cmap=cmap, 
                square=True,
                linewidth=.5, cbar_kws={"shrink": .5}, ax=ax)
    else:
        sns.heatmap(df, cmap=cmap, 
                square=True,
                linewidth=.5, cbar_kws={"shrink": .5}, ax=ax)


CorrMtx(corr, dropDuplicates = True)
```
