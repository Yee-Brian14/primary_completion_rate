From above $r^2$ data, we see that this really isn't a good linear relationship. However, there's definitely a relationship here -- P > |t| is less than .05! So, we want to standardize and transform data to see how that can help us. Standardizing won't help with the model, but it will help us interpret results later on. 

```python
# standardize the data
father_support_df = data[['father_support', 'adolescent_birth_rate', 'Average_Urban_Population']].copy(deep=True)

abr_mean = father_support_df['adolescent_birth_rate'].mean()
abr_std = father_support_df['adolescent_birth_rate'].std()

aup_mean = father_support_df['Average_Urban_Population'].mean()
aup_std = father_support_df['Average_Urban_Population'].std()

father_support_df['abr'] = father_support_df['adolescent_birth_rate'].apply(lambda x: (x-abr_mean)/abr_std)
father_support_df['aup'] = father_support_df['Average_Urban_Population'].apply(lambda x: (x-aup_mean)/aup_std)
```
