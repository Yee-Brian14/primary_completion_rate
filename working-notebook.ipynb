```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20,10))

N = 7

# count for each of income groups, by region
low = np.array(data.groupby('Region').IncomeGroup.value_counts().unstack().fillna(0)['Low income'])
lower_mid = np.array(data.groupby('Region').IncomeGroup.value_counts().unstack().fillna(0)['Lower middle income'])
upper_mid = np.array(data.groupby('Region').IncomeGroup.value_counts().unstack().fillna(0)['Upper middle income'])
high = np.array(data.groupby('Region').IncomeGroup.value_counts().unstack().fillna(0)['High income'])

ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

# plot each income type
p1 = plt.bar(ind, high, width,bottom=low+lower_mid+upper_mid)
p2 = plt.bar(ind, upper_mid, width,bottom=low+lower_mid)
p3 = plt.bar(ind, lower_mid, width,bottom=low)
p4 = plt.bar(ind, low, width)

plt.ylabel('Income')
plt.title('Income by Region')
plt.xticks(ind, ('East Asia & Pacific', 'Europe & Central Asia', 'Latin America & Caribbean', 
                 'Middle East & North Africa', 'North America', 'South Asia', 'Sub-Saharan Africa'))

plt.yticks(np.arange(0, 80, 10))
plt.legend((p1[0], p2[0], p3[0], p4[0]), ('High', 'Upper Middle', 'Lower Middle', 'Low'))

plt.show()
```

```python
fig, ax = plt.subplots(1,7,figsize=(70,10))
#data.groupby(['Region', 'relative_country_income'])[['Average Primary School Completion']].count()
colors = ['red','blue','green','black','orange','purple','pink']
count = 0
for region in list(zip(data['Region'].unique(),colors)):
    ax[count].scatter(x=data[(data['Region'] == region[0])][['relative_country_income']],
                     y=data[(data['Region'] == region[0])][['Average Primary School Completion']],
                     color = region[1], label=region[0], s=100)
    ax[count].set_title(region[0],size=30)
    ax[count].set_xlabel('Income')
    ax[count].set_ylabel('Primary Completion')
    ax[count].set_ylim([40,140])
    count += 1

plt.suptitle('Income vs Primary Completion by Region',size=40)
plt.show()
```
